```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler

iris = load_iris()
X = iris['data']
y = iris['target']
names = iris['target_names']
feature_names = iris['feature_names']

# One hot encoding
enc = OneHotEncoder()
Y = enc.fit_transform(y[:, np.newaxis]).toarray()

# Scale data to have mean 0 and variance 1

# which is importance for convergence of the neural network
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data set into training and testing
X_train, X_test, Y_train, Y_test = train_test_split(
    X_scaled, y, test_size=0.5, random_state=2)

n_features = X.shape[1]
n_classes = Y.shape[1]
```

```python
input = 4
ouput = 3
roll_no = 4

hidden_layer = roll_no % 5 + 3
hidden_node = hidden_layer + random.randint(1,5)

print(hidden_layer, hidden_node)
```

```python
def create_model(input_dim, output_dim, h_node, h_layer, name):
  model = Sequential(name=name)

  for i in range(h_layer):
    model.add(Dense(h_node,activation='relu',input_dim=input_dim))
  model.add(Dense(output_dim, activation='softmax'))

  model.compile(optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy'])

  return model

#neural network.
models = [create_model(4, 3, hidden_node, i, 'model_{}'.format(i-3)) for i in range(hidden_layer,hidden_layer+3)]

for model in models:
  model.summary()
```

```python
history_dict = {}

for model in models:
  trained_model = model.fit(X_train,Y_train, epochs = 50, verbose=0,validation_data=(X_test, Y_test))

  score = model.evaluate(X_test, Y_test, verbose=0)
  print(model.name)
  print('Test loss:', score[0])
  print('Test accuracy:', score[1])
  print('-----------------------------------------')

  history_dict[model.name] = trained_model
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
for model_name in history_dict:
  plt.plot(history_dict[model_name].history['val_accuracy'],label=model_name)
  plt.xlabel('epochs')
  plt.ylabel('val_accuracy')
plt.legend()

plt.subplot(1,2,2)
for model_name in history_dict:
  plt.plot(history_dict[model_name].history['val_loss'],label=model_name)
  plt.xlabel('epochs')
  plt.ylabel('val_loss')
plt.legend()
```
